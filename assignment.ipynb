```
import pandas as pd
from pandasql import sqldf
import plotly.express as px

pysqldf = lambda q: sqldf(q, globals())

file_path = "Data Engineer Task Assignment.xlsx"

messages = pd.read_excel(file_path, sheet_name="Messages")
statuses = pd.read_excel(file_path, sheet_name="Statuses")

messages.head()
```

```
statuses.head()
```

```
messages.shape
statuses.shape
```

```
messages.columns
statuses.columns
```

```
# Convert messages date columns
date_cols_messages = [
    "external_timestamp",
    "last_status_timestamp",
    "inserted_at",
    "updated_at"
]

for col in date_cols_messages:
    if col in messages.columns:
        messages[col] = pd.to_datetime(messages[col])

# Convert statuses date columns
date_cols_status = [
    "timestamp",
    "inserted_at",
    "updated_at"
]

for col in date_cols_status:
    if col in statuses.columns:
        statuses[col] = pd.to_datetime(statuses[col])
```

```
messages_dedup = pysqldf("""
SELECT *
FROM messages m1
WHERE updated_at = (
    SELECT MAX(updated_at)
    FROM messages m2
    WHERE m1.uuid = m2.uuid
)
""")

messages_dedup.head()
```

```
len(messages), len(messages_dedup)
```

```
status_agg = pysqldf("""
SELECT
    message_uuid,
    MIN(CASE WHEN status = 'sent' THEN timestamp END) AS sent_ts,
    MIN(CASE WHEN status = 'delivered' THEN timestamp END) AS delivered_ts,
    MIN(CASE WHEN status = 'read' THEN timestamp END) AS read_ts,
    MIN(CASE WHEN status = 'failed' THEN timestamp END) AS failed_ts
FROM statuses
GROUP BY message_uuid
""")

status_agg.head()
```

```
message_fact = pysqldf("""
SELECT
    m.uuid AS message_uuid,
    m.direction,
    CASE 
        WHEN m.direction = 'inbound' THEN m.masked_from_addr
        ELSE m.masked_addressees
    END AS user_id,
    m.content,
    m.rendered_content,
    m.inserted_at,
    m.last_status,
    s.sent_ts,
    s.delivered_ts,
    s.read_ts,
    s.failed_ts
FROM messages_dedup m
LEFT JOIN status_agg s
ON m.uuid = s.message_uuid
""")

message_fact.head()
```

```
pysqldf("""
SELECT message_uuid, COUNT(*) as cnt
FROM message_fact
GROUP BY message_uuid
HAVING COUNT(*) > 1
""")
```

```
pysqldf("""
SELECT *
FROM message_fact
WHERE read_ts < sent_ts
""")
```

```
pysqldf("""
SELECT 
    COUNT(*) as violation_count
FROM message_fact
WHERE (read_ts < sent_ts)
   OR (delivered_ts < sent_ts)
""")
```

```
pysqldf("""
SELECT COUNT(*) as no_status_messages
FROM message_fact
WHERE sent_ts IS NULL
AND delivered_ts IS NULL
AND read_ts IS NULL
AND failed_ts IS NULL
""")
```

```
pysqldf("""
SELECT direction, COUNT(*) as cnt
FROM message_fact
WHERE sent_ts IS NULL
AND delivered_ts IS NULL
AND read_ts IS NULL
AND failed_ts IS NULL
GROUP BY direction
""")
```

```
pysqldf("""
SELECT message_uuid, last_status
FROM message_fact
WHERE direction='outbound'
AND sent_ts IS NULL
""")
```

```
pysqldf("""
SELECT COUNT(*) as total_outbound
FROM message_fact
WHERE direction='outbound'
""")
```

```
weekly = pysqldf("""
SELECT
    strftime('%Y-%W', inserted_at) AS week,
    COUNT(DISTINCT user_id) AS total_users,
    COUNT(DISTINCT CASE WHEN direction='inbound' THEN user_id END) AS active_users
FROM message_fact
GROUP BY week
ORDER BY week
""")

weekly.head()
```

```
import plotly.express as px

fig = px.line(
    weekly,
    x="week",
    y=["total_users", "active_users"],
    title="Total vs Active Users Over Time"
)

fig.show()
```

```
read_fraction = pysqldf("""
SELECT 
    CAST(SUM(CASE WHEN read_ts IS NOT NULL THEN 1 ELSE 0 END) AS FLOAT)
    /
    COUNT(*) AS read_fraction
FROM message_fact
WHERE direction='outbound'
AND failed_ts IS NULL
""")

read_fraction
```

```
rf = read_fraction.iloc[0,0]

import plotly.express as px

px.bar(
    x=["Read Rate"],
    y=[rf],
    title="Read Rate of Non-Failed Outbound Messages"
)
```

```
time_dist = pysqldf("""
SELECT 
    (julianday(read_ts) - julianday(sent_ts))*24*60 AS minutes_to_read
FROM message_fact
WHERE direction='outbound'
AND read_ts IS NOT NULL
AND failed_ts IS NULL
""")

time_dist.head()
```

```
px.histogram(
    time_dist,
    x="minutes_to_read",
    nbins=40,
    title="Distribution of Time to Read (Minutes)"
)
```

```
time_dist_filtered = pysqldf("""
SELECT 
    (julianday(read_ts) - julianday(sent_ts))*24*60 AS minutes_to_read
FROM message_fact
WHERE direction='outbound'
AND read_ts IS NOT NULL
AND failed_ts IS NULL
AND (julianday(read_ts) - julianday(sent_ts))*24*60 < 1440
""")

px.histogram(
    time_dist_filtered,
    x="minutes_to_read",
    nbins=40,
    title="Time to Read (Within 24 Hours)"
)
```

```
last_week = pysqldf("""
SELECT last_status, COUNT(*) as cnt
FROM message_fact
WHERE direction='outbound'
AND inserted_at >= date('now','-7 day')
GROUP BY last_status
""")

last_week
```

```
last_week = pysqldf("""
SELECT last_status, COUNT(*) as cnt
FROM message_fact
WHERE direction='outbound'
AND inserted_at >= (
    SELECT datetime(MAX(inserted_at), '-7 days')
    FROM message_fact
)
GROUP BY last_status
""")

last_week
```

```
px.bar(
    last_week,
    x="last_status",
    y="cnt",
    title="Outbound Messages by Status (Last 7 Days in Dataset)"
)
```
